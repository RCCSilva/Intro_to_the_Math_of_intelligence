```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

First, we start define our function that gets a $x$ as input and gives us $y$ as output.

$$ y = m \times x  + b $$

Second, for this function, we'll define a loss functions which gives us how far are our predictions from the observed values.

$$ \sum_{i = 0}^{N} (y - \hat{y}) ^ {2}$$ 

Following StatQuest video, let's start taking the partial derivative for the intercept ($b$) as if $m = 1.47$ and let's say we only have the first three data points and the initial $b$ is 0. 

```python
points = np.genfromtxt('data.csv', delimiter=',')
dataset_3 = points[:3]
dataset_3
```

```python
plt.title('Iteration 0')
plt.scatter(dataset_3[:,0], dataset_3[:,1]);
x = np.linspace(0,80,1000)
y = 1.47 * x + 0
plt.plot(x,y);
```

Oukay. Now if we plug our values inside the sum of squares function, we'll get:

$$ SumOfSquares = (31.70 - (32.50 \times 1.47) + b)^{2} + (68.77 - (53.42 \times 1.47) + b)^{2} + (62.56 - (61.53 \times 1.47) + b)^{2}$$

Gradient descent takes the partial derivative of the SumOfSquares and updates the value of $m$ and $b$.

$$\frac{d}{d b} SumOfSquares = \frac{d}{d b} (31.70 - (32.50 \times 1.47) + b)^{2} + \frac{d}{d b}(68.77 - (53.42 \times 1.47) + b)^{2} + \frac{d}{d b}(62.56 - (61.53 \times 1.47) + b)^{2}$$

Let's start with the first value. 

$$ \frac{d}{d b} (31.70 - (32.50 \times 1.47 + b))^{2}$$

Using the __chain rule__.


$$ \frac{d}{d b} 31.70 - 32.50 \times 1.47 + (-1)b = -1$$

Thus,

$$ \frac{d}{d b} (31.70 - (32.50 \times 1.47 + b))^{2} = -2(31.70 - (32.50 \times 1.47 + b))$$

We can replicate this process for the other two parts and we'll get:

$$\frac{d}{d b} SumOfSquares = (-2)(31.70 - (32.50 \times 1.47) + b) + (-2)(68.77 - (53.42 \times 1.47) + b) + (-2)(62.56 - (61.53 \times 1.47) + b)$$

```python
intercept_gradient = 0
initial_b = 0
m = 1.47
learning_rate = 0.01
for i in range(len(dataset_3)):
    x = dataset_3[i,0]
    y = dataset_3[i,1]
    intercept_gradient += (-2) * (y - ((1.47 * x) + initial_b))
    
step_size = intercept_gradient * learning_rate
new_b = initial_b - step_size
print('Step size: {}'.format(step_size))
print('New intercept: {}'.format(new_b))
```

```python
plt.title('Initial State')
plt.scatter(dataset_3[:,0], dataset_3[:,1]);
x = np.linspace(0,80,1000)
y = 1.47 * x + new_b
plt.plot(x,y);
```

Let's iterate for 1000 times.

```python
intercept_gradient = 0
new_b = 0
m = 1.47
learning_rate = 0.01
for i in range(1000):
    for k in range(len(dataset_3)):
        x = dataset_3[k,0]
        y = dataset_3[k,1]
        intercept_gradient += (-2) * (y - ((1.47 * x) + new_b))
    
    step_size = intercept_gradient * learning_rate
    new_b = initial_b - step_size
    
plt.title('Iteration {}'.format(i + 1))
plt.scatter(dataset_3[:,0], dataset_3[:,1]);
x = np.linspace(0,80,1000)
y = 1.47 * x + new_b
plt.plot(x,y);

print('Intercept: {:.2f}'.format(new_b))
```

HOW COOL IS THAT?!

After 1000 iterations, we obtained $b=17.91$ using gradient descent.

## Gradient Descent: Slope and Intercept

Now it's time to take the gradient descent for both the __slope__ and the __intercept__. The process is quite similar. We'll take the partial derivative for the __intercept__ and after for __slope__.

### Intercept

$$ \frac{d}{d b} (31.70 - (32.50 \times m + b))^{2}$$

Using the __chain rule__:

$$ \frac{d}{d b} 31.70 - 32.50 \times m + (-1)b = -1$$

Thus,

$$ \frac{d}{d b} (31.70 - (32.50 \times m + b))^{2} = -2(31.70 - (32.50 \times m + b))$$

We can replicate this process for the other two parts and we'll get:

$$\frac{d}{d b} SumOfSquares = (-2)(31.70 - (32.50 \times m) + b) + (-2)(68.77 - (53.42 \times m) + b) + (-2)(62.56 - (61.53 \times m) + b)$$

### Slope

$$ \frac{d}{d m} (31.70 - (32.50 \times m + b))^{2}$$

Using the __chain rule__ again:

$$ \frac{d}{d m} 31.70 - (32.50 \times m + b) = - 32.50$$

$$ \frac{d}{d m} (-2)(32.50)  \times (31.70 - (32.50 m + b))^{2}$$

And we apply the same equation for the other parts.

$$\frac{d}{d b} SumOfSquares = (-2)(32.50) \times (31.70 - (32.50 \times m) + b) + (-2)(53.42) \times (68.77 - (53.42 \times m) + b) + (-2)(61.53) \times (62.56 - (61.53 \times m) + b)$$

```python
def step_gradient(current_m, current_b, points, learning_rate):
    gradient_m = 0
    gradient_b = 0
    N = float(len(points))
    for i in range(len(points)):
        x = points[i,0]
        y = points[i,1]
        gradient_m += (-2/N) * (x) * (y - ((x * current_m) + current_b))
        gradient_b += (-2/N) * (y - ((x * current_m) + current_b))
        
    new_m = current_m - (gradient_m * learning_rate)
    
    new_b = current_b - (gradient_b * learning_rate)
    
    return new_m, new_b

def gradient_runner(points, n_iterations, initial_m, initial_b, learning_rate):
    m = initial_m
    b = initial_b
    for _ in range(n_iterations):
        m, b = step_gradient(m, b, points, learning_rate)
    return m, b

learning_rate = 0.00001
initial_b = 0
initial_m = 0
n_iterations = 1000
points = np.genfromtxt('data.csv', delimiter=',')

m, b = gradient_runner(points, n_iterations, initial_m, initial_b, learning_rate)
print('Slope: {:.2f}\nIntercept: {:.2f}'.format(m,b))
plt.scatter(points[:,0], points[:,1]);
x = np.linspace(0,100,1000)
y = m * x + b
plt.plot(x,y);
```
